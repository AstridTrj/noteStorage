```python
import os
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.common.exceptions import NoSuchElementException


class Spider:
    def __init__(self, delay=30):
        firefox_profile = webdriver.FirefoxProfile()
        # 禁止加载图片
        firefox_profile.set_preference('permissions.default.image', 2)
        # 禁用css
        firefox_profile.set_preference('permissions.default.stylesheet', 2)
        # 禁用js
        firefox_profile.set_preference('javascript.enabled', 'false')
        self.browser = webdriver.Firefox(firefox_profile=firefox_profile)
        self.wait = WebDriverWait(self.browser, delay)
        self.browser.maximize_window()
        self.main()

    def load_file(self):
        file_path_1 = "城市_全部.txt"
        result_ls = []
        with open(file_path_1, encoding="utf-8") as fp:
            for city in fp.readlines():
                if city.strip():
                    for year in range(2003, 2020):
                        result_ls.append([city.strip(), year, ""])
        # 加载爬取结果
        file_path_2 = "result.csv"
        ls = []
        if os.path.exists(file_path_2):
            with open(file_path_2, encoding="gbk") as fp:
                for line in csv.reader(fp):
                    if line and line[2] not in ["", "0"]:  # 过滤掉专利数量为空或者为0的记录
                        line = list(line)
                        line[1] = int(line[1])
                        line[2] = int(line[2])
                        ls.append(line)
        # 合并爬取结果
        for line in ls:
            self.insert(line, result_ls)
        self.save_file(result_ls)
        return result_ls

    def main(self):
        start_url = "http://epub.cnipa.gov.cn/gjcx.jsp"
        result_ls = self.load_file()
        file_path = "城市_待爬.txt"
        city_ls = []
        with open(file_path, encoding="utf-8") as fp:
            for i in fp.readlines():
                if i.strip():
                    city_ls.append(i.strip())

        to_do_list = []
        for city in city_ls:
            for year in range(2003, 2020):
                to_do_list.append([city, year])

        # 过滤已爬取的记录
        i = 0
        while i < len(to_do_list):
            for line in result_ls:
                if to_do_list[i][0] == line[0] and to_do_list[i][1] == line[1] and line[2] and line[2] > 0:
                    to_do_list.remove(to_do_list[i])
                    i -= 1
                    break
            i += 1

        retry = {}  # 记录重复次数
        while to_do_list:
            tmp = to_do_list.pop(0)
            city = tmp[0]
            year = tmp[1]
            try:
                self.browser.get(start_url)
                bt = self.clickable('//img[@src="images/js_btn.jpg"]')
                self.browser.find_element_by_xpath('//input[@id="ad1"]').send_keys("{}.01.01".format(year))
                self.browser.find_element_by_xpath('//input[@id="ad2"]').send_keys("{}.12.31".format(year))
                self.browser.find_element_by_xpath('//input[@id="paa"]').send_keys(city)
                time.sleep(1)
                self.browser.execute_script("arguments[0].scrollIntoView();", bt)  # 滑动页面，定位到某一位置
                bt.click()
                self.clickable('//li[@class="lb_a"]').click()
                self.wait.until(
                    ec.presence_of_element_located((By.XPATH, '//table'))
                )

                if self.element_exist('//div[@class="next"]'):
                    self.browser.find_element_by_xpath('//div[@class="next"]/a[last()-1]').click()
                    self.wait.until(
                        ec.text_to_be_present_in_element((By.XPATH, '//div[@class="next"]/a[1]'), "<")
                    )
                num = int(self.browser.find_element_by_xpath('//td//tr[last()]/td').text)

                self.insert([city, year, num], result_ls)
                print("{}, {}, {}".format(city, year, num))
                self.save_file(result_ls)
            except:
                s = "{}{}".format(city, year)
                retry[s] = retry.get(s, 0) + 1
                if retry[s] <= 5:  # 最多重爬5次
                    to_do_list.insert(0, tmp)
                else:
                    self.insert([city, year, 0], result_ls)
                    print("{}, {}, {}: doubt".format(city, year, 0))
            finally:
                self.browser.delete_all_cookies()

    def clickable(self, xpath_str):
        return self.wait.until(
            ec.element_to_be_clickable((By.XPATH, xpath_str))
        )

    def element_exist(self, xpath_str):
        try:
            return self.browser.find_element_by_xpath(xpath_str)
        except NoSuchElementException:
            return None

    @staticmethod
    def save_file(result_ls):
        save_path = "./result.csv"
        with open(save_path, "w", encoding="gbk", newline='') as fp:
            writer = csv.writer(fp, delimiter=",")
            writer.writerows(result_ls)

    @staticmethod
    def insert(param, result_ls):
        for i in range(len(result_ls)):
            if result_ls[i][0] == param[0] and result_ls[i][1] == param[1]:
                result_ls[i][2] = param[2]
                break


if __name__ == '__main__':
    spider = Spider()
```
