```python
import torch
```

```python
x = torch.ones(5, 3)
if torch.cuda.is_available:
    print("yes")
    device = torch.device("cuda")
    y = torch.ones_like(x, device=device)
    x = x.to(device)
    z = x + y
    print(z)
    print(z.to("cpu"))
```

```python
n, model_in, hid, model_out, epochs = 1000, 128, 512, 10, 500
x = torch.randn(n, model_in)
y = torch.randn(n, model_out)
learning_rate = 1e-3
```

```python
class TwoLayerNN(torch.nn.Module):
    def __init__(self, m_in, hid, m_out):
        super(TwoLayerNN, self).__init__()
        self.linear1 = torch.nn.Linear(m_in, hid)
        self.linear2 = torch.nn.Linear(hid, m_out)
    
    def forward(self, x):
        out = self.linear1(x)
        out = out.clamp(min=0)
        out = self.linear2(out)
        return out

    
model = TwoLayerNN(model_in, hid, model_out)
loss_function = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

for epo in range(epochs):
    model.train()
    optimizer.zero_grad()
    y_pred = model(x)
    loss = loss_function(y, y_pred)
    print(epo + 1, "  ", loss.item())
    loss.backward()
    optimizer.step()
```
